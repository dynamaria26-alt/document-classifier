```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline

# Sample training data
documents = [
    "India won the cricket match",
    "Virat Kohli scored a century",
    "The stock market is rising",
    "Investors are buying shares",
    "Artificial Intelligence is changing technology",
    "Python is popular for machine learning"
]

labels = [
    "Sports",
    "Sports",
    "Business",
    "Business",
    "Technology",
    "Technology"
]

# Create classifier
model = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('clf', MultinomialNB())
])

# Train model
model.fit(documents, labels)

# Test
text = ["Machine learning and AI are growing rapidly"]
prediction = model.predict(text)

print("Category:", prediction[0])
```

```python
while True:

    text = input("Enter document: ")

    if text.lower() == "exit":
        break

    prediction = model.predict([text])

    print("Predicted Category:", prediction[0])
```
